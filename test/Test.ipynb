```python
%matplotlib inline
import csv
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
files = ['ahash.txt', 'dhash.txt', 'phash.txt', 'whash.txt']
avg_speed = []
for file in files:
    file_speed = []
    
    with open(file) as f:
        data=[tuple(line) for line in csv.reader(f)]
        data = [t for t in data if t != ()]
        
    for i in range(11):
        avg = 0
        for j in range(i*5, i*5+5):
            avg+= float(data[j][1])
            
        avg /= 5
        file_speed.append(avg)
        
    avg_speed.append(file_speed)       
```

```python
labels = ['aHash', 'dHash', 'pHash', 'wHash']
y = list(range(11))

sns.set_style("darkgrid")
plt.figure(figsize=(10,10))
plt.title("Runtime")
plt.xlabel("Time(seconds)")
plt.ylabel("Max Hamming Distance")

for speed in avg_speed:
    plt.plot(speed, y)
plt.legend(labels)
plt.show()
```

```python
true_positives = []
false_positives = []
with open("changes.txt") as f:
    actual = [tuple(line) for line in csv.reader(f)]
    actual = dict([t for t in actual if t != ()])
       
for i in range(11):
    tp = 0
    fp = 0
    filepath = "./ahash/foundchanges" + str(i*5) + ".txt"
    with open(filepath) as f:
        data=[tuple(line) for line in csv.reader(f)]
        data = [t for t in data if t != ()]

    for pair in data:
        if pair[0] in actual:
            if actual[pair[0]] == pair[1]:
                tp += 1
            else:
                fp += 1
        elif pair[1] in actual:
            if actual[pair[1]] == pair[0]:
                    tp += 1
            else:
                fp += 1
        else:
            fp += 1
                
            
    true_positives.append(tp)
    false_positives.append(fp)

print(list(zip(true_positives, false_positives)))
```

```python
labels = ['True positives', 'False Positives']
y = list(range(11))

sns.set_style("darkgrid")
plt.figure(figsize=(10,10))
plt.title("aHash True Positives vs. False Positives")
plt.xlabel("Number")
plt.ylabel("Max Hamming Distance")
plt.plot(true_positives, y)
plt.plot(false_positives, y)
plt.legend(labels)
plt.show()
```
